```python
A = [1, 3, 5, 7, 4]

sum=0
for num in A :
  if num % 2 != 0:
    sum = sum +num
print(sum)
```

```python
A = [1, 3, 5, 7, 4]
sum = 0
for num in range(len(A)):
  if num % 2 != 0:
    sum = sum + A[num]
print(sum)
```

```python
A = [1, 3, 5, 7, 4]
odd_count=0
even_count=0
for num in A :
  if num % 2 != 0:
    odd_count= odd_count+1
  else:
    even_count=even_count+1
print(f'Number of odd:{odd_count}')
print(f'Number od even:{even_count}')

```
